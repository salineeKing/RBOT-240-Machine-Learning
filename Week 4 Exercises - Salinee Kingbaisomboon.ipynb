**(Exercise D1)** What is an activation function?  Why use them?

<font color="blue"><u><b>Answer</b></u></font>

Activation functions are mathematical equations that determine the output of a neural network. The function is attached to each neuron in the network, and determines whether it should be activated ("fired") or not, based on whether each neuron’s input is relevant for the model’s prediction. Activation functions also help normalize the output of each neuron to a range between 1 and 0 or between -1 and 1.[1]

The purpose of the activation function is to introduce non-linearity into the output of a neuron. 

Non-linearity is needed in activation functions because its aim in a neural network is to produce a nonlinear decision boundary via non-linear combinations of the weight and inputs.

**<u>Answer for question 5 (Programming Exercise 6.2.1):</u>** 

Construct an image X with diagonal edges.
* What happens if you apply the kernel K in this section to it?
* What happens if you transpose X?
* What happens if you transpose K?

<font color="blue"><u><b>Answer</b></u></font>

**What is Diagonal?** [1]

A diagonal is a straight line connecting the opposite corners of a  through its . 

<image src="https://cdn-skill.splashmath.com/panel-uploads/GlossaryTerm/d36231360bc440ccac86b71c26c39e4a/1559998365_diagonal-shapes.png" />

[1] https://www.splashlearn.com/math-vocabulary/geometry/diagonal

```
# Import Library
import torch
from torch import nn

# Define funtion to accept an input tensor X and a kernel tensor K and returns an output tensor Y
def corr2d(X, K):
    """Compute 2D cross-correlation."""
    h, w = K.shape
    Y = torch.zeros((X.shape[0] - h + 1, X.shape[1] - w + 1))
    for i in range(Y.shape[0]):
        for j in range(Y.shape[1]):
            Y[i, j] = (X[i:i + h, j:j + w] * K).sum()
    return Y

# Construct an image "X" of  6×8  pixels with diagonal edges with a convention of black in 0 and white in 1
X = torch.ones(8)
X_diag = torch.diag(X) # Reference of how to use torch.diag ==> https://pytorch.org/docs/stable/generated/torch.diag.html
print('\033[34m' + '\033[1m' + 'Image X with diagonal edges' + '\033[0m')
print(X_diag)

# Define the kernel K in this section
K = torch.tensor([[1.0, -1.0]])

# Apply the kernel K in this section to it
print('\033[34m' + '\033[1m' + 'Apply the kernel K in this section to it' + '\033[0m')
Y = corr2d(X_diag, K)
print(Y)

# Transpose X
print('\033[34m' + '\033[1m' + 'Transpose X' + '\033[0m')
Y_X_trans = corr2d(X_diag.t(), K)
print(Y_X_trans)

# Transpose K
print('\033[34m' + '\033[1m' + 'Transpose K' + '\033[0m')
Y_K_trans = corr2d(X_diag, K.t())
print(Y_K_trans)
```

<font color="green"><b>Answer:</b></font> The function will detect -1 for the diagonal edge from black to white from left bottom corner and 1 for the diagonal edge from white to black from right top corner. All other outputs take value 0.
* **What happens if you transpose X?** 

<font color="green"><b>Answer:</b></font> No change result from the non-transpose of X.
* **What happens if you transpose K?**

<font color="green"><b>Answer:</b></font> The function will detect 1 for the diagonal edge from white to black from left bottom corner and -1 for the diagonal edge from black to white from right top corner. All other outputs take value 0. The result is an opposite side of the non-transpose X.

**(Exercise P1)** Pytorch includes many [pre-defined models](https://pytorch.org/docs/stable/torchvision/models.html).  Try performing the KITTI classification using a pre-defined model.

<font color="blue"><u><b>Answer</b></u></font>

In this exercise, I choose **AlexNet** model which is one of the most popular convolutional neural network model used in deep learning. In the following code, I will just use the default pre-defined model from PyTorch without any additional traning. 

Also, I've download imagenet classes from [here](https://github.com/deep-learning-with-pytorch/dlwpt-code/blob/master/data/p1ch2/imagenet_classes.txt) to use in a code as a pre-defined labels correspondent with **AlexNet** model.


```
# Import Library
from google.colab import drive
import torch
from torch import nn
import torchvision
from torchvision import models
from torchvision import transforms
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
import json
import collections

from google.colab.patches import cv2_imshow

# Mount Google Drive
drive.mount('/content/gdrive')
```

```
# Defined functions used with KITTI dataset
class KittiDataset(torch.utils.data.Dataset):
  def __init__(self, dataset_path, transform=None):
    self.images = np.load(os.path.join(dataset_path, 'images.npy'), mmap_mode='r')

    with open(os.path.join(dataset_path, 'metadata.json')) as metadata_file:
        self.metadata = json.load(metadata_file)

    self.transform = transform

  def __getitem__(self, index):    
    sample = {
      'image': self.images[index],
      'metadata': self.metadata[index]
    }

    if self.transform:
        sample = self.transform(sample)
        
    return sample

  def __len__(self):
      return len(self.images)

def collate(batch):
  images = torch.tensor(np.stack([sample['image'] for sample in batch]))
  metadata = [sample['metadata'] for sample in batch]
  return {'images': images, 'metadata': metadata}
```

```
from torch.utils import data
import torch.optim as optim

# Load KITTI dataset
dataset = KittiDataset('gdrive/MyDrive/Colab Notebooks/')
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [int(.8*len(dataset)), int(.2*len(dataset))])

train_dataloader = data.DataLoader(
    train_dataset,
    num_workers=1,
    batch_size=64,
    collate_fn=collate
)

test_dataloader = data.DataLoader(
    test_dataset,
    num_workers=1,
    batch_size=64,
    collate_fn=collate
)
```

```
# Load AlexNet pre-defined class labels
with open('gdrive/MyDrive/Colab Notebooks/imagenet_classes.txt') as f:
  class_labels = [line.strip() for line in f.readlines()]
```

```
object_types = ['Pedestrian', 'Truck', 'Cyclist', 'Car', 'Misc', 'Van', 'Tram', 'Person_sitting']

#Function to show some random images
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)).astype('uint8'))
    plt.show()

#Get some random testing images
dataiter = iter(test_dataloader)
datalist = dataiter.next()
images, label_list = datalist['images'], [object_types.index(item['label']) for item in datalist['metadata']]

#Show images (only first five)
imshow(torchvision.utils.make_grid(images[:5]))
# print labels
print(' '.join('%5s' % object_types[label_list[j]] for j in range(5)))
```

```
# Use Pre-trained model
alexnet = models.alexnet(pretrained=True)
alexnet.eval()
```

```
# Use AlexNet to predict images from KITTI dataset
with torch.no_grad():
    for data in images[:5]: # Process only first five images
      input_tensor = torch.unsqueeze(data, 0)
      prediction_tensor = alexnet(input_tensor.type(torch.FloatTensor))

      max_value, index_of_max_value = torch.max(prediction_tensor, 1)
      print(index_of_max_value.numpy())

      predicted_label = class_labels[index_of_max_value]

      npimg = data.numpy()
      plt.imshow(np.transpose(npimg, (1, 2, 0)).astype('uint8'))
      plt.show()

      print(predicted_label)
```

We can see that the predictions from the KITTI dataset didn't do a very good job. All these images are incorrectly predicted.

We will try using the more clearly picture of the **car** without any interferenced with the background scenario. (The **car** will be on focus).

```
# Define image preprocessor
preprocess_image = transforms.Compose([
  transforms.Resize(256),
  transforms.CenterCrop(224),
  transforms.ToTensor(),
  transforms.Normalize(
  mean=[0.485, 0.456, 0.406],
  std=[0.229, 0.224, 0.225]
)])

# Use AlexNet with a better image
image = Image.open("gdrive/MyDrive/Colab Notebooks/car.jpg")
image_tensor = preprocess_image(image)
input_tensor = torch.unsqueeze(image_tensor, 0)
prediction_tensor = alexnet(input_tensor)
max_value, index_of_max_value = torch.max(prediction_tensor, 1)
predicted_label = class_labels[index_of_max_value]
cv2_imshow(np.array(image))
print(predicted_label)
```

We can see that using the image with a well-focus object will yeild a better result. The image was predicted with a correct label.

The reason might be bacause of the original pre-trained **AlextNet** model was primary used to detect many types of object such as cat, dog, boat and such. It's for a general purpose, not for detecting objects on the streets.

To be able to use **AlextNet** with KITTI dataset, we need to train the model especially so it will detect the cars and object on the streets better.
