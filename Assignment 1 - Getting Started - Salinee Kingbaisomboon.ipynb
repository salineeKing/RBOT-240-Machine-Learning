 ### A Brief Side Note on Tensors
 
You may be wondering at this point "what is a tensor?".  If you've got a programming background, the best way to think of it is as a multidimensional array.  That is to say, it's simply a bunch of numbers arranged into some regular grid.  The number of indicies for the multidimensional array is commonly refered to as the "rank" or "order" of the tensor.  So, vectors are an example of a rank 1 tensor, and matricies are examples of rank 2 tensors.  Rank 3 tensors are commonly used to store images, because they require an array of shape  color_channels x height x width, which I'll abbreviate (c, h, w).  In machine learning, we often pass a "batch" of multiple images through the network, so the data is then stored in a rank 4 tensor of shape (b, c, h, w)

  Now, there is also a mathematical meaning to the word tensor as well which you would find if you were to look up what a tensor is on Wikipedia, but this meaning is not terribly relevant for what we'll be covering in this class.

## Probability
Next we're going to look at using PyTorch to explore some of the concepts in probability you've been reading in the learning material.  Since probability is so important to machine learning, PyTorch contains the [distributions](https://pytorch.org/docs/stable/distributions.html) package entirely dedicated for dealing with probability distributions.

```
import numpy as np
import torch
import math
import matplotlib.pyplot as plt
```

```
# Use the PyTorch distributions package to generate 10000 samples from 
# a Gaussian (I.E. Normal) distribution with a mean of 5 and std of 2
# Store the result in a variable called "gaussian_samples"

gaussian_samples = torch.normal(mean=5.0, std=2.0, size=(1, 10000))
```

```
# This should look like the familiar Gaussian curve
plt.hist(gaussian_samples, bins=100)
plt.title("Histogram of 10,000 Samples generated from Gaussian Distribution")
plt.show()
```

## Below is a Gaussian function formula
<img src='https://wikimedia.org/api/rest_v1/media/math/render/svg/61463c3decedda46e356782e24051ec7dd3c34c8' />

*Reference: https://en.wikipedia.org/wiki/Gaussian_function*

```
# Create variables called "mu" and "sigma" that contain the mean and standard
# deviation of gaussian_samples, respectively

mu = gaussian_samples.mean()
sigma = gaussian_samples.std()

# Write a function called gaussian() that takes a PyTorch tensor of numbers
# and computes the value of the Gaussian function for each value

def gaussian(tensor):
  gaussian_function = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(tensor - mu)**2 / (2 * sigma**2))
  return gaussian_function

# # Let's compare the histogram of the samples vs the gaussian function.  
# # These should match up fairly well
x = torch.linspace(0, 10, 100)
plt.hist(gaussian_samples, bins=100)
plt.plot(x, gaussian(x)*10000/(2*math.pi))
plt.title("Probability Density of Gaussian Distrbution")
plt.show()
```

```
# Lets explore drawing various numbers of samples from a Uniform 
# distribution to see what the histogram of the sums look like.  

# Use the PyTorch distributions package to create a distribution object that 
# generates values uniformly between 0 and 100

# sampler_uniform = torch.distributions.uniform.Uniform(torch.Tensor([0.0]),torch.Tensor([100.0])) // This code doesn't work and it will take a long time
sampler_uniform = torch.distributions.uniform.Uniform(0.0, 100.0);

# We want to do experiments with summing various numbers of samples 
# from this distribution, ranging from summing 1 to 5 samples

# variable with size of 5 to hold the sum of num_samples per each dimension
# samples = torch.zeros(5)

# Include the end (from 1 to 5)
for num_samples in range(1,6):
  # sample a 10000 x num_samples tensor from the distribution and find the
  # sum along the "num_samples" dimension (so we have the sum from 
  # 10000 experiments)

  result = sampler_uniform.sample((10000,num_samples))
  # dimension = num_samples-1
  # samples[dimension] = torch.sum(result) # [10,000,1:dimension]

  samples = result.sum(dim=1);

  # You should see a series of plots with the first looking fairly uniform,
  # but then looking progressively more Gaussian-like.  This is a consequence
  # of the Central Limit Theorem
  
  plt.figure()
  plt.hist(samples, bins=100)
  plt.title("Histogram of sum of uniformly distribution by " + str(num_samples) + " samples for over 10,000 experiments")
  plt.show()
```
