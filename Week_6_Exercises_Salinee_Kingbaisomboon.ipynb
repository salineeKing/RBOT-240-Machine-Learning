##### <u><b>Markov Chain</b></u>
A Markov chain essentially consists of a set of transitions, which are determined by some probability distribution, that satisfy the Markov property.

##### <u><b>Hidde Markov Chain (HMM)</b></u>

Ok. Things get better in the **HMM**. In the HMM, we are constructing an inference model based on the assumptions of a Markov process that <font color="green"><b>the future is independent of the past given that we know the present. (future state is related to the immediately previous state and not the states before that)</b></font>

Please look at the following picture to get a better idea:

<image src="https://miro.medium.com/max/700/1*icuG9VNHLItN50cMdBJDbA.png" />

In **HMM**, we observe the outputs over time to determine the sequence based on how likely they were to produce that output.

##### <u><b>Linear Dynamical System (LDS)</b></u>
Dynamical Systems are systems, described by one or more equations, that evolve over time. The **linear dynamical system (LDS)** model is arguably the most commonly used time series model for real-world engineering and financial applications due to its relative simplicity, mathematically predictable behavior, and the fact that exact inference and predictions for the model can be done efficiently.

**Hidden Markov Models (HMMs) and Linear Dynamical Systems (LDSs) are based on the same assumption: a hidden state variable.** The only different is that the HMM uses a discrete state variable with arbitrary dynamics and arbitrary measurements while the LDS uses a continuous state variable with linearGaussian dynamics and measurements.[3]

```
################################################
####### Place your implementation here #########
################################################

# Import Library
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import numpy as np

# Create bayes network in python using pgmpy
bayesNet = BayesianModel()

# Add Nodes from exercise 8.3
bayesNet.add_node("C")
bayesNet.add_node("A")
bayesNet.add_node("B")

# Define the Edge
bayesNet.add_edge("C", "A")
bayesNet.add_edge("C", "B")

# Add CPDs for each node (the numbers came from proof of exercise 8.3)
cpd_C = TabularCPD('C', 2, values=[[0.48], [.520]]) # C = 0, 1 respectively

cpd_A = TabularCPD('A', 2, values=[[0.500, 0.692], [0.500, 0.308]], # C = 0 & A = 0, C = 1 & A = 0, C = 0 & A = 1, C = 1 & A = 1 respectively
                   evidence=['C'], evidence_card=[2])

cpd_B = TabularCPD('B', 2,
                   values=[[0.800, 0.400], [0.200, 0.600]], # C = 0 & B = 0, C = 1 & B = 0, C = 0 & B = 1, C = 1 & B = 1 respectively
                   evidence=['C'], evidence_card=[2])
bayesNet.add_cpds(cpd_C, cpd_A, cpd_B)

bayesNet.check_model()
print("\nModel is correct.")

solver = VariableElimination(bayesNet)

# CPT table of C
result = solver.query(variables=['C'])
print("C", result)

# CPT table of A
result = solver.query(variables=['A'])
print("A", result)

# CPT table of B
result = solver.query(variables=['B'])
print("B", result)

# CPT table of A,B,C
result = solver.query(variables=['A', 'B', 'C'])
print("A, B, C", result)

# Show dependencies between the variables
bayesNet.get_independencies()
```
