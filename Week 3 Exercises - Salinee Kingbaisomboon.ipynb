**(Exercise 4.12)** Verify the relation (4.88) for the derivative of the logistic sigmoid function defined by (4.59).

<font color="blue"><u><b>Answer</b></u></font>

We're given below equation in the question:

$
\begin{align}
\mathbf{\frac{dσ(a)}{da}} = σ(1-σ)
\tag*{(4.88)}
\end{align}
$

In the Bishop's textbox, below is a  ***logistic sigmoid*** function which is defined by:

$
\begin{align}
\mathbf{σ({a})} = \frac{1}{1 + \exp(-a)}
\tag*{(4.59)}
\end{align}
$



$
\begin{align}
\text{Differentiate an equation above with regard to a:}
\end{align}
$

\begin{equation}
\begin{aligned}
\mathbf{\frac{dσ(a)}{da}} &= \frac{exp(a)}{(1+exp(-a))^2} \\
  & = \frac{exp(a)}{1+exp(-a)} \cdot \frac{1}{1+exp(-a)}\\
  & = \frac{1}{1+exp(-a)} \cdot \frac{exp(a)}{1+exp(-a)}\\
  & = σ({a})\left( \frac{exp(-a)}{1+exp(-a)}\right)\\
  & = σ({a})\left( \frac{1+exp(-a)}{1+exp(-a)} - \frac{1}{1+exp(-a)}\right)\\
  & = σ({a})\left( 1 - σ({a})\right)\\
\end{aligned}
\end{equation}

Above expression is equal to ***(4.88)***

**(Exercise D2)** In assignment 2, we'll look at using the normal equations, gradient descent, and stochastic gradient descent to find the optimal parameters for a linear model.  How do these approaches compare in terms of computational cost?

<font color="blue"><u><b>Answer</b></u></font>

Amoung these three approaches, we can say that **stochastic gradient descent** is outperform the rest approaches in term of computational cost.

**Stochastic gradient descent** (SGD or "on-line") typically reaches convergence much faster than batch (or "standard") gradient descent since it updates weight more frequently. 

Unlike the gradient descent which computes the gradient using the whole dataset, because the SGD, also known as incremental gradient descent, tries to find minimums or maximums by iteration from a single randomly picked training example

Below are the advantages which support why choosing **Stochastic gradient descent** is better in term of the computational cost[1]:

1.   It is easier to fit into memory due to a single training sample being processed by the network.
2.   It is computationally fast as only one sample is processed at a time.
3.   Since it uodate the parameters more frequently, we will see a significant decrease in a computational time, especially for the larger dataset.
4. Due to frequent updates the steps taken towards the minima of the loss function have oscillations which can help getting out of local minimums of the loss function (in case the computed position turns out to be the local minimum).

Based on this post from Medium, ***Why gradient descent and normal equation are BAD for linear regression***[2], the post stated that using both **Normal equation** and **Gradient descent** won't recommened since both of them are slow. There are many reasons to support that, for example, iteratively approaching the minimum takes quite a bit of time (gradient descent) or due to its Algorithm complexity (Normal equation).

Based on this website, we can see that:[3]
 
*   The Algorithm complexity of **gradient descent** is $O(kn^2)$
*   The Algorithm complexity of **normal equation** is $O(n^3)$

Remark: $n$ is the number of features

The higher Complexity of an algorithm will require a very long time to process the data. Especially, if the number of features are very large, it will suffer from the curse of dimension.

However, **stochastic gradient descent** is a bit different than those batch algorithm. The reason is that in terms of the complexity is constant (linear).

**(Exercise P4)** A fantastic Python package for quickly using many of these models in a black-box manner is [scikit-learn](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression), including both [linear regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression) and [logistic regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression).

Use scikit-learn to repeat linear regression experiments in assignment 2.

```
## Answer
## Below is the complete code using scikit-learn library in order to repeat linear regression experiments

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Paramters of generated data (same with Assignment 2)
num_samples = 50

x_val = -3 + 6*np.random.rand(num_samples)
y_val = np.sin(x_val) + .2*np.random.rand(num_samples) 

# Put data in dataframe
simulate_data = pd.DataFrame({'x': x_val, 'y':y_val})

# Print DataFrame's head
print(simulate_data.head())

# sckit-learn implementation

# Create numpy arrays for the x (independent) and y (dependent) variables
x = simulate_data.x.to_numpy().reshape(-1,1)
y = simulate_data.y.to_numpy().reshape(-1,1)

# Model initialization
regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(x, y)
# Predict
y_predicted = regression_model.predict(x)

# Model statistic evaluation
rmse = mean_squared_error(y, y_predicted)

# Plotting values
# data points
plt.scatter(x, y, s=20)
plt.xlabel('x')
plt.ylabel('y')
# predicted values
plt.plot(x, y_predicted, color='r')
plt.title('x vs y and its Linear Regression line')
plt.show()

# Printing statistic evaluation values
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
```
