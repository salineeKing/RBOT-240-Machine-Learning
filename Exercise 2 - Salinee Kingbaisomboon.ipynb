In this question, I will implement the **Multiplicative Inverse of a Matrix[1]** which stated that:

When $A$ is multiplied by $A^{-1}$ the result is the identity matrix $\Rightarrow$ $AA^{-1} = I$

We are given the inverse of the precision matrix formula:

$
\begin{align}
\mathbf{R} = \begin{pmatrix}

$

\mathbf{

A-BD^{-1}C\\

We need to show that the inverse of the precision matrix ***(2.104)*** is given by the covariance matrix ***(2.105)*** which has the following formula:

$
\begin{align}
\mathbf{cov[z]} =
R^{-1} =
 \begin{pmatrix}
Λ^{-1}&Λ^{-1}A^{T}\\
AΛ^{-1}&L^{-1}+AΛ^{-1}A^T\\
\end{pmatrix}
\tag*{(2.104)}
\end{align}
$

By comparing element-wise between equation ***(2.76)*** and ***(2.104)***
, we clearly see that, by comparing $\mathbf{R}$ to ***(2.76)***:

1.   **$A = Λ+A^TLA$**
2.   **$B = -A^TL$**
3.   **$C = -LA$**
4.   **$D = L$**

We will perform the matrix operation on equation ***(2.76)*** by substitute values of each element from above as follow:

1.   **$M$ (Top Left)**

  $
  \begin{align}
  \text{By substitute $A$, $B$, $C$ and $D$}
  \end{align}
  $

  \begin{equation}
  \begin{aligned}
  \mathbf{M} &= \left(A-BD^{-1}C\right)^{-1} \\
  & = \left((Λ+A^TLA)-(-A^TL)(L^{-1})(-LA)\right)^{-1} \\
  & = \left((Λ+A^TLA)-(-A^TLL^{-1})(-LA)\right)^{-1} \\
  & = \left((Λ+A^TLA)-(-A^TI)(-LA)\right)^{-1} \\
  & = \left((Λ+A^TLA)-(A^TLA)\right)^{-1} \\

  \end{equation}

2.   **$-MBD^{-1}$ (Top Right)**

  \text{By substitute $M$ (from previous solving), $B$ and $D$}
  \end{align}

  & = -Λ^{-1}(-A^TLL^{-1}) \\

  & = Λ^{-1}A^T\\

  \end{equation}

3.   **$-D^{-1}CM$ (Bottom Left)**

  \begin{align}

  & = (-L^{-1})(-L)(A)(Λ^{-1}) \\
  & = I(A)(Λ^{-1}) \\
  & = AΛ^{-1}\\

  \end{equation}

4.   **$D^{-1}+D^{-1}CMBD^{-1}$ (Bottom Right)**

  & = (L^{-1})+ (IAΛ^{-1}I) \\
  & = L^{-1}+ AΛ^{-1}

  \end{equation}

Putting them all together, we can see that the new metrix is the same as  covariance matrix ***(2.104)***$\Rightarrow$

**Answer for question 2 (Exercise 2.39):** 

We are given below equations for the posterior distribution of the mean of a Gaussian random variable:

$
\begin{align}
\mathbf{μ_{N}} = \frac{σ^2}{Nσ_{0}^2+σ^2}μ_{0} + \frac{Nσ_{0}^2}{Nσ_{0}^2+σ^2}μ_{ML}
\tag*{(2.141)}
\end{align}
$

$
\begin{align}
\mathbf{\frac{1}{σ_{N}^2}} = \frac{1}{σ_{0}^2} + \frac{N}{σ^2}
\tag*{(2.142)}

$
\begin{align}
\mathbf{μ_{ML}} = \frac{1}{N}\sum_{n=1}^{N} x_{n}
\tag*{(2.143)}
\end{align}
$


We can see that from the ***(2.141)***, we can eliminate the denominator on both side of an equation if we multiply by ***(2.142)***.

$
\begin{align}
\text{Therefore,}
\end{align}
$

\begin{equation}
\begin{aligned}
\mathbf{\frac{μ_{N}}{σ_{N}^2}} &= \left(\frac{1}{σ_{0}^2} + \frac{N}{σ^2}\right)\left(\frac{σ^2}{Nσ_{0}^2+σ^2}μ_{0} + \frac{Nσ_{0}^2}{Nσ_{0}^2+σ^2}μ_{ML}\right) \\
  & = \left(\frac{σ^2+Nσ_{0}^2}{σ_{0}^2σ^2}\right)\left(\frac{σ^2}{Nσ_{0}^2+σ^2}μ_{0} + \frac{Nσ_{0}^2}{Nσ_{0}^2+σ^2}μ_{ML}\right) \\
  & = \frac{μ_{0}}{σ_{0}^2}+\frac{Nμ_{ML}}{σ^2} \\
  & = \frac{μ_{0}}{σ_{0}^2} + \frac{N\sum_{n=1}^{N} x_{n}}{Nσ^2}\\
  & = \frac{μ_{0}}{σ_{0}^2} + \frac{\sum_{n=1}^{N-1} x_{n}}{σ^2} + \frac{x_{N}}{σ^2}\\
  & = \left(\frac{μ_{0}}{σ_{0}^2} + \frac{\sum_{n=1}^{N-1} x_{n}}{σ^2}\right)+ \frac{x_{N}}{σ^2} \\
  & = \frac{μ_{N-1}}{σ_{N-1}^2}+\frac{x_{N}}{σ^2}
\end{aligned}
\end{equation}

Above expression is an expressions for the sequential update of **$μ_{N}$** and **$σ^2$** from the first **$N-1$**.

**Answer for question 3 (Exercise 3.17):** 

We need to prove that the the evidence function for the Bayesian linear regression model can be written in the form ***(3.78)*** which is:

$
\begin{align}
\mathbf{p(t|\alpha,\beta)} = \left(\frac{\beta}{2\pi}\right)^\frac{N}{2} \left(\frac{\alpha}{2\pi}\right)^\frac{M}{2} \int \exp \{-E(w)\} dw
\tag*{(3.78)}
\end{align}
$

We are given below equations to evaluate the **Evidence Function**:

$
\begin{align}
\mathbf{\ln p(t|w,\beta)} = \frac{N}{2}\ln\beta - \frac{N}{2}\ln(2\pi) - \beta E_{D}(w)
\tag*{(3.11)}
\end{align}
$


$
\begin{align}
\mathbf{E_{D}(w)} = \frac{1}{2}\sum_{n=1}^{N}\{t_{n} - w^T\phi(x_{n})\}^2
\tag*{(3.12)}
\end{align}
$

$
\begin{align}
\mathbf{p(t|\alpha,\beta)} = \int p(t|w,\beta) p(w|\alpha) dw
\tag*{(3.77)}
\end{align}
$

We will also referencing below equations from **Chapter 1 - Introduction** of the textbook in order to solve this question:

$
\begin{align}
\mathbf{N(x|μ,σ^2)} = \frac{1}{(2\piσ^2)^{1/2}} \exp\{-\frac{1}{2σ^2}(x-μ)^2\}
\tag*{(1.46)}
\end{align}
$

$
\begin{align}
\mathbf{p(w|\alpha)} = N(w|0,\alpha^{-1}I) = \left(\frac{\alpha}{2\pi}\right)^\frac{M+1}{2}\exp\{-\frac{\alpha}{2}w^Tw\}
\tag*{(1.65)}
\end{align}
$

We will start from equation ***(3.77)*** by rewriting the first two portion of the equation.

$\text{By utilizing the form of the Gaussian distribution, given by (1.46) but with the ∏ (product-based form)}$

\begin{equation}
\begin{aligned}
\mathbf{p(t|w,\beta)} &= \prod_{n = 1}^{N}\frac{1}{(2\pi\beta^{-1})^{1/2}} \exp\{-\frac{1}{2\beta^{-1}}(t_{n}-\phi(x_{n})^Tw)^2\} \\
  & = \left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\exp\{\sum_{n=1}^{N} -\frac{\beta}{2}(t_{n}-\phi(x_{n})^Tw)^2 \} \\
  & = \left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\exp\{-\frac{\beta}{2} ||t-\phi w||^2\}
\end{aligned}
\end{equation}
By pluging an above expression together with ***(1.65)*** into ***(3.77)***:

\begin{equation}
\begin{aligned}
\mathbf{p(t|\alpha,\beta)} &= \int p(t|w,\beta) p(w|\alpha) dw \\
  & = \int \bigg[ \left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\exp\{-\frac{\beta}{2} ||t-\phi w||^2\} \bigg] \bigg[ \left(\frac{\alpha}{2\pi}\right)^\frac{M+1}{2}\exp\{-\frac{\alpha}{2}w^Tw\} \bigg] dw \\
  & = \bigg[\left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\int \exp\{-\frac{\beta}{2} ||t-\phi w||^2\} \bigg] \bigg[ \left(\frac{\alpha}{2\pi}\right)^\frac{M+1}{2} \int \exp\{-\frac{\alpha}{2}||w||^2\} \bigg] dw\\
  & = \bigg[\left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\left(\frac{\alpha}{2\pi}\right)^\frac{M+1}{2} \bigg]\bigg[ \int \exp\{-\frac{\beta}{2} ||t-\phi w||^2\}\int \exp\{-\frac{\alpha}{2}||w||^2\}\bigg] dw\\
  & = \bigg[\left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\left(\frac{\alpha}{2\pi}\right)^\frac{M}{2} \bigg]\bigg[ \int \exp\{\left(-\frac{\beta}{2} ||t-\phi w||^2\right)\left(-\frac{1}{2}||w||^2\right)\}\bigg] dw\\
  & = \bigg[\left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\left(\frac{\alpha}{2\pi}\right)^\frac{M}{2} \bigg]-\beta\bigg[ \int \exp\{\left(-\frac{1}{2} ||t-\phi w||^2\right)\left(-\frac{1}{2}||w||^2\right)\}\bigg] dw\\
  & = \bigg[\left(\frac{\beta}{2\pi}\right)^{\frac{N}{2}}\left(\frac{\alpha}{2\pi}\right)^\frac{M}{2} \bigg]-\beta E_{D}(w)
\end{aligned}
\end{equation}

We will get an expression above which is equal to ***(3.78)***
