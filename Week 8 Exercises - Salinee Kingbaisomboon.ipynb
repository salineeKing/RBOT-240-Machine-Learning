**<font color="blue">(Exercise PRML 12.3)** Verify that the eigenvectors defined by (12.30) are normalized to unit length, assuming that the eigenvectors $\textbf{v}_i$ have unit length.</font>

<font color="green"><u><b>Answer</b></u></font>

We're given below equation in the question (based on Bishop's textbook):

$
\begin{align}
\mathbf{\textbf{u}_i} = \frac{1}{(N\lambda_{i})^{1/2}} X^{T}\textbf{v}_i
\tag*{(12.30)}
\end{align}
$

Based on equations **(12.28)** from Bishop's textbook:

$
\begin{align}
\mathbf{\frac{1}{N}XX^{T}\textbf{v}_i} = \lambda_{i}\textbf{v}_i\
\tag*{(12.28)}
\end{align}
$

From **(12.30)**, we can obtained the following:

\begin{equation}
\begin{aligned}
\mathbf{\textbf{u}^{T}_{i} \cdot \textbf{u}_i} &= \big(\frac{1}{(N\lambda_{i})^{1/2}} X^{T}\textbf{v}_i \big)^{T} \cdot \big( \frac{1}{(N\lambda_{i})^{1/2}} X^{T}\textbf{v}_i\big)\\
  & = \frac{1}{(N\lambda_{i})} \cdot \textbf{v}^{T}_iX \cdot X^{T}\textbf{v}_i\\
  & = \frac{1}{(N\lambda_{i})} \textbf{v}^{T}_iX X^{T}\textbf{v}_i\
\text{(Equation: a)}
\end{aligned}
\end{equation}

From **(12.28)**, we left multiply $\textbf{v}^{T}_{i}$ and devide $\lambda_{i}$ on both sides of the equation to yeild:

\begin{equation}
\begin{aligned}
\mathbf{\frac{1}{N\lambda_{i}}\textbf{v}^{T}_{i} XX^{T}\textbf{v}_i} &= \textbf{v}^{T}_{i}\cdot \textbf{v}_i \\
& = \lVert \textbf{v}_i \rVert^2 \
\text{(Equation: b)}
\end{aligned}
\end{equation}

Substitute equation **a** into an equation **b**:

\begin{equation}
\begin{aligned}
\mathbf{\textbf{u}^{T}_{i} \cdot \textbf{u}_i} &= \lVert \textbf{v}_i \rVert^2 \\
\end{aligned}
\end{equation}

We have used the fact that $\textbf{v}_i$ has unit length.

\begin{equation}
\begin{aligned}
\text{Therefore, }
\mathbf{\textbf{u}^{T}_{i} \cdot \textbf{u}_i} &= 1 \\
\end{aligned}
\end{equation}

<font color="green"><u><b>Answer</b></u></font>

##### <u><b>What is eigendecomposition?</b></u>

The eigendecomposition is one form of matrix decomposition. Decomposing a matrix means that we want to find a product of matrices that is equal to the initial matrix. In the case of eigendecomposition, we decompose the initial matrix into the product of its eigenvectors and eigenvalues.

The very well known usage of **Eigen** is to use it as **Transformation matrix** which the transformed matrix is more suitable for computation.[1]

Please see below image of displaying an affine transformation in Eigen:

<image src="https://wiki.pathmind.com/images/wiki/two_spaces.png" />

A matrix is decomposed into:


1.   A diagonal matrix formed from **eigenvalues** of matrix-A
2.   And a matrix formed by the **eigenvectors** of matrix-A


##### <u><b>Use of eigendecomposition for PCA analysis</b></u>

In **PCA**, **eigendecomposition** is used as an element to simplify the calculation of other more complex matrix operations. 

It often happens that your data are presented to you provided with many features, sometimes hundreds of them but we don't need all of them for the analysis.

Principal components are new variables that are constructed as linear combinations of the initial variables. These combinations are done in such a way that these new variables are uncorrelated and most of the information within the initial variables is stored into the first components. One of the popular method for this transformation was done via **eigendecomposition**.[2]

##### <u><b> How is it used for PCA?</b></u>

The eigenvectors and eigenvalues of a covariance (or correlation) matrix represent the "core" of a PCA: The eigenvectors (principal components) determine the directions of the new feature space, and the eigenvalues determine their magnitude.

***Covariance matrix*** (also called as dispersion matrix or variance-covariance matrix) is a matrix whose element in the i,j position is the covariance between the i-th and j-th element (feature) of a random vector (A random vector is a random variable with multiple dimensions).

***Eigenvectors*** are unit vectors with length or magnitude equal to 1. They are often referred to as right vectors, which simply means a column vector.

***Eigenvalues*** are coefficients applied to eigenvectors that give the vectors their length or magnitude.

So, Eigendecompistion used in PCA as the following:[3]

*   Measures how each variable is associated with one another using a Covariance matrix.
*   Understands the directions of the spread of our data using Eigenvectors.
*   Brings out the relative importance of these directions using Eigenvalues.

```
# Make some data
import torch

torch.manual_seed(0)

X = torch.randn(100, 2) @ torch.tensor([[40.0,20.0], [-20.0, 1.0]])

# Convert to Numpy Array
X = X.numpy()
```

**<font color="blue">(Exercise P1)** Use eigenvalue decomposition to calculate the principal component of the data.  Show the principal component on the plot, either as an arrow, or by projecting the data points to it.</font>

*(hint: look at Appendix B of the tutorial reading)*

<font color="green"><u><b>Answer</b></u></font>



```
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore") # To suppress warning

%matplotlib inline

plt.figure()
plt.scatter(X[:,0],X[:,1])

################################################
####### Place your implementation here #########
################################################

# Statndadization and Deduct Mean to make the data centered
X = StandardScaler().fit_transform(X)
X -= np.mean(X, axis=0)

n, m = X.shape

# Covariance matrix
C = np.cov(X, rowvar=False)
# C = (X - X.mean(0)).T.dot((X - X.mean(0))) / (X.shape[0]-1)

# C is a symmetric matrix and so it can be diagonalized:
eigenvalues, eigenvectors = np.linalg.eig(C)

# Sort eigenvalues result
idx = eigenvalues.argsort()[::-1]
eigenvalues, eigenvectors = eigenvalues[idx], eigenvectors[:, idx]

# Vector to plot
exp_var = eigenvalues
components = eigenvectors

print('\033[34m' + '\033[1m' + 'Explained Variance' + '\033[0m')
print(exp_var)
print('\033[34m' + '\033[1m' + 'Principal Components' + '\033[0m')
print(components)

v1 = 2 * components[0] * np.sqrt(exp_var[0])
v2 = 2 * components[1] * np.sqrt(exp_var[1])
c = (0, 0) # Center is at 0,0 because of our standardization


plt.annotate('', c + v1, c, arrowprops={'arrowstyle': '->', 'shrinkA': 0,
                                        'shrinkB': 0, 'linewidth': 3})
plt.annotate('', c + v2, c, arrowprops={'arrowstyle': '->', 'shrinkA': 0,
                                        'shrinkB': 0, 'linewidth': 3})

plt.axis('equal')
plt.grid()
plt.title('PCA with Eigenvalue Decomposition')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal');
plt.show()
```
